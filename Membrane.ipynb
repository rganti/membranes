```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from plot_objects import Plot
from tabulated_energies import Potential
from matplotlib import rc
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (8.0, 6.0)
```

**Tail-Tail interactions (Attractive)**

```python
sigma = 1
w_c = 1.4

tail_tail = Potential(sigma, w_c)
r_full, force_tail_tail, v_tail_tail = tail_tail.attractive()
```

```python
plt_v_tt = Plot(r_full, v_tail_tail, "\mathrm{Energy\;Tail-Tail}")
plt_f_tt = Plot(r_full, force_tail_tail, "\mathrm{Force\;Tail-Tail}")
plt_f_tt.plot_f()
plt_v_tt.plot_f(ylo=-2,yhi=2, xlo=0.5, xhi=3)
```

**Monomer-Tail interactions (Attractive)**

```python
sigma = 1
mon = Potential(sigma, w_c)
r_full, force_mon_tail, v_mon_tail = mon.attractive()
```

```python
plt_v_mt = Plot(r_full, v_mon_tail, "\mathrm{Energy\;Monomer-Tail}")
plt_f_mt = Plot(r_full, force_mon_tail, "\mathrm{Force\;Monomer-Tail}")
plt_f_mt.plot_f()
plt_v_mt.plot_f(ylo=-2,yhi=2, xlo=0.5, xhi=3)
```

**Monomer-Head interactions (WCA)**

```python
r_full, force_mon_head, v_mon_head = mon.repulsive()
```

```python
plt_v_mh = Plot(r_full, v_mon_head, "\mathrm{Energy\;Monomer-Head}")
plt_f_mh = Plot(r_full, force_mon_head, "\mathrm{Force\;Monomer-Head}")
plt_f_mh.plot_f()
plt_v_mh.plot_f(ylo=-1,yhi=2, xlo=0.5, xhi=3)
```

**Head-Head and Head-Tail interactions (WCA)**

```python
sigma = 0.95
head_tail = Potential(sigma, w_c)
r_full, force_head_tail, v_head_tail = head_tail.repulsive()
```

```python
plt_f_hh = Plot(r_full, force_head_tail, "\mathrm{Force\;Head-Tail, Head-Head}")
plt_f_hh.plot_f()

plt_v_hh = Plot(r_full, v_head_tail, "\mathrm{Energy\;Head-Tail, Head-Head}")
plt_v_hh.plot_f(ylo=-1,yhi=2, xlo=0.5, xhi=3)
```

**Monomer-Monomer interactions (None)**

```python
sigma = 2
mon_mon = Potential(sigma, w_c)
r_full, force_mon_mon, v_mon_mon = mon_mon.attractive()
force_mon_mon.fill(0)
v_mon_mon.fill(0)
```

```python
plt_v_mm = Plot(r_full, v_mon_mon, "\mathrm{Energy\;Monomer-Monomer}")
plt_f_mm = Plot(r_full, force_mon_mon, "\mathrm{Force\;Monomer-Monomer}")
plt_f_mm.plot_f()
plt_v_mm.plot_f(ylo=-1,yhi=2, xlo=0.5, xhi=3)
```
